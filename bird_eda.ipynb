```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
# Read in dataframe of diets

diet_df = pd.read_csv('BirdFuncDat.txt', sep=None, engine='python')
```

```python
# Read in dataframe of threats

df_threats_habitats = pd.read_csv('threats_habitats.csv')

df_threats_habitats = df_threats_habitats.rename(columns={'Scientific name': 'Scientific'})
```

```python
# Test merge on scientific

scientific_merge = diet_df.merge(df_threats_habitats, how='left', left_on='Scientific', right_on='Scientific')
```

```python
# Test merge on english name

english_merge = diet_df.merge(df_threats_habitats, how='left', left_on='English', right_on='English name')

english_merge = english_merge.rename(columns={'Scientific_x': 'Scientific'})

english_merge.drop('Scientific_y', axis=1, inplace=True)
```

```python
# Concatenate two datasets (stack on top of each other), remove nulls, then remove duplicates

to_concat = [english_merge, scientific_merge]

combined_df = pd.concat(to_concat)

df_clean = combined_df.dropna(subset=['biological_threat'], axis=0)

df_clean = df_clean.drop_duplicates(subset='Scientific')
```
